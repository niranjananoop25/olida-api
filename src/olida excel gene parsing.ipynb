# Import Modules

```python
import pandas as pd
```

# Get Excel Data

```python
# Get gene data
df_gene = pd.read_excel('data/Gene.xlsx', skiprows=1)

# Get combo data
df_combo = pd.read_excel('data/Combination.xlsx', skiprows=1)

# Get disease data
df_disease = pd.read_excel('data/Disease.xlsx', skiprows=1)

#Get gene combo data
df_gene_combo = pd.read_excel('data/GeneCombination.xlsx', skiprows=1)
```

```python
df_disease.sort_values(by='Combinations')
```

# One-to-one mapping: Genes to Combos and Diseases

```python
df_gene['Combinations'] = df_gene['Combinations'].str.replace(';','')
df_gene['Combinations'] = df_gene['Combinations'].apply(str.split)
df_gene = df_gene.explode('Combinations')
df_gene
```

```python
df_gene.sort_values(by = 'Combinations')
```

```python
df_gene.drop(columns=['Ensembl Id', 'Entrez Id', 'Chromosome', 'Uniprot Accession Number', 'Gene Ontology Molecular Function', 'Essential In Mouse', 'Pathway', 'Variants'], inplace=True)
```

```python
df_gene
```

```python
df_gene.sort_values(by='Combinations')
```

# One-to-one mapping: Combo to Disease ICD

```python
df_disease['Combinations'] = df_disease['Combinations'].str.replace(';','')
df_disease['Combinations'] = df_disease['Combinations'].apply(str.split)
df_disease = df_disease.explode('Combinations')
df_disease
```

```python
df_disease.sort_values(by = 'Combinations')
```

```python
df_disease.drop(columns = ['Orphanet ID', 'Disease Name', 'ICD-10 Category', 'Omim Id'], inplace=True)
```

```python
df_disease
```

# One-to-one mapping: Combos to Partner Genes

```python
df_gene_combo
```

```python
df_gene_combo.drop(columns=['Entry Id', 'Genes Relationship', 'Protein Interactions', 'Common Pathways', 'Common Pathways', 'GENEmeta'], inplace=True)
df_gene_combo
```

```python
df_gene_combo['Oligogenic variant combinations'] = df_gene_combo['Oligogenic variant combinations'].str.replace(';','')
df_gene_combo['Oligogenic variant combinations'] = df_gene_combo['Oligogenic variant combinations'].apply(str.split)
df_gene_combo = df_gene_combo.explode('Oligogenic variant combinations')
df_gene_combo
```

```python
df_gene_combo.rename(columns = {'Oligogenic variant combinations':'Combinations'}, inplace = True)
df_gene_combo
```
